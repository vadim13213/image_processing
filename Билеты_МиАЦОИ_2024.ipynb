```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/images (1).jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# метода ближайшего соседа
image_nearest = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_NEAREST)
image_nearest_rgb = cv2.cvtColor(image_nearest, cv2.COLOR_BGR2RGB)

# билинейная интерполяция
image_bilinear = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
image_bilinear_rgb = cv2.cvtColor(image_bilinear, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 7))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Оригинальная фотка')
plt.axis('off')

# Метод ближайшего соседа
plt.subplot(1, 3, 2)
plt.imshow(image_nearest_rgb)
plt.title('Ближайший сосед')
plt.axis('off')

# Билинейная интерполяция
plt.subplot(1, 3, 3)
plt.imshow(image_bilinear_rgb)
plt.title('Билинейная интерполяция')
plt.axis('off')

plt.show()
```

```python
#2 Применить морфологическую операцию «закрытие»* с разными структурными элементами, сравнить результат.

image = cv2.imread('/content/images (1).jpg', cv2.IMREAD_GRAYSCALE) #гружу в чб

# маленький квадрат
kernel_square = np.ones((5, 5), np.uint8)

# крест
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

# эллипс
kernel_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

# операция закрытия (дилатация + эрозия)
closing_square = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel_square)
closing_cross = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel_cross)
closing_ellipse = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel_ellipse)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Оригинальная чб фотка')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(closing_square, cmap='gray')
plt.title('Квадратный элемент')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(closing_cross, cmap='gray')
plt.title('Крестовидный элемент')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(closing_ellipse, cmap='gray')
plt.title('Эллиптический элемент')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/images (1).jpg', cv2.IMREAD_GRAYSCALE)

# эрозии
kernel = np.ones((5, 5), np.uint8)
eroded_image = cv2.erode(image, kernel, iterations=1)

edge_by_subtraction = cv2.subtract(image, eroded_image)

_, binary_edge_by_subtraction = cv2.threshold(edge_by_subtraction, 50, 255, cv2.THRESH_BINARY)

edges_canny = cv2.Canny(image, 100, 200)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_edge_by_subtraction, cmap='gray')
plt.title('Границы (вычитание эрозированного, БИНАРИЗАЦИЯ)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges_canny, cmap='gray')
plt.title('Границы (Canny)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
