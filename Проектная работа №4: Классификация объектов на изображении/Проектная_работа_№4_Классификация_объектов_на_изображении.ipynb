```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image):
    """
    Преобразует изображение в оттенки серого, применяет медианное размытие,
    адаптивный порог и оператор Кэнни для улучшенной предобработки.
    """
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Преобразование в оттенки серого
    blurred = cv2.medianBlur(gray, 5)              # Медианное размытие для устранения шума
    adaptive_thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                            cv2.THRESH_BINARY, 11, 2)  # Адаптивный порог
    edges = cv2.Canny(adaptive_thresh, 30, 150)    # Обнаружение краев

    # Морфологические операции
    kernel = np.ones((3, 3), np.uint8)
    edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)  # Закрытие (устранение разрывов)
    edges = cv2.morphologyEx(edges, cv2.MORPH_OPEN, kernel)   # Открытие (удаление шума)

    return edges

def detect_coins(edges):
    """
    Обнаруживает круги на изображении с использованием преобразования Хафа.
    """
    circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=1.3, minDist=30,
                               param1=100, param2=30, minRadius=15, maxRadius=80)
    if circles is not None:
        circles = np.uint16(np.around(circles))
        return circles[0, :]
    return []

def refine_circles(circles, edges):
    """
    Уточняет радиусы и фильтрует круги по форме (округлость).
    """
    refined_circles = []
    for circle in circles:
        x, y, radius = circle
        # Проверка округлости: площадь контура близка к площади круга
        mask = np.zeros_like(edges)
        cv2.circle(mask, (x, y), radius, 255, -1)
        contour, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if contour:
            area = cv2.contourArea(contour[0])
            circularity = 4 * np.pi * (area / (2 * np.pi * radius) ** 2)
            if 0.7 <= circularity <= 1.3:  # Фильтрация по округлости
                refined_circles.append(circle)
    return np.array(refined_circles)

def classify_coins(circles):
    """
    Определяет номинал монет на основе радиуса и возвращает общую сумму.
    """
    total_sum = 0
    coin_values = []

    for circle in circles:
        radius = circle[2]
        if radius < 20:
            value = 1  # 1 рубль
        elif radius < 30:
            value = 2  # 2 рубля
        elif radius < 45:
            value = 5  # 5 рублей
        else:
            value = 10 # 10 рублей
        total_sum += value
        coin_values.append(value)

    return total_sum, coin_values

def draw_results(image, circles, coin_values, total_sum):
    """
    Рисует обнаруженные круги и подписывает номиналы на изображении.
    """
    for i, circle in enumerate(circles):
        x, y, radius = circle
        value = coin_values[i]
        cv2.circle(image, (x, y), radius, (0, 255, 0), 2)  # Окружность
        cv2.putText(image, f'{value}R', (x - 20, y + 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)  # Текст

    cv2.putText(image, f'Total: {total_sum} RUB', (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)  # Общая сумма

    return image

def display_image(image, title="Image"):
    """
    Отображает изображение в Google Colab.
    """
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

def main():
    """
    Основная функция программы.
    """
    # Загрузка изображения
    from google.colab import files
    uploaded = files.upload()
    if not uploaded:
        print("Ошибка: изображение не загружено.")
        return

    image_path = list(uploaded.keys())[0]
    image = cv2.imread(image_path)
    if image is None:
        print("Ошибка: изображение не найдено.")
        return

    # Предварительная обработка
    edges = preprocess_image(image)

    # Визуализация промежуточных краев
    display_image(cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), title="Обнаруженные границы")

    # Обнаружение монет
    circles = detect_coins(edges)
    if len(circles) == 0:
        print("Монеты не обнаружены.")
        return

    # Уточнение кругов
    circles = refine_circles(circles, edges)

    # Классификация монет
    total_sum, coin_values = classify_coins(circles)

    # Отображение результатов
    result_image = draw_results(image.copy(), circles, coin_values, total_sum)
    display_image(result_image, title="Обработанное изображение")

if __name__ == "__main__":
    main()
```
